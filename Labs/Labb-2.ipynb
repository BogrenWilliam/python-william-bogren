```python
# Huvuduppgift
import re
import numpy as np
import math
import matplotlib.pyplot as plt

with open("../data/datapoints.txt") as f_read:
    data_points = f_read.readlines()  
    data_points.pop(0) # "pop" returns the removed value and 1st line with (0)

    PokemonWidth = []
    PokemonHeigth = []
    PokemonLabel = []


    
    for line in data_points: # Changes the text and places in three elements, for each step
        

        Width, Heigth, Label = [float(1) for i in line.split (",")]

        PokemonWidth.append(Width)
        PokemonHeigth.append(Heigth)
        PokemonLabel.append(Label)




def Pokemon_plot(): 

    for (Widthplot, Heigthplot, Labelplot) in zip(PokemonWidth, PokemonHeigth,PokemonLabel): 
        if Labelplot == 0: 
            color_of_Pokemon = "Red"
        else: 
            color_of_Pokemon = "Blue"
        plt.plot(Widthplot, Heigthplot, Labelplot,"o", markerfacecolor=color_of_Pokemon, markersize = 5)
plt.grid()
    
    



Pokemon_plot()

with open ("../Data/testpoints.txt") as file:
    test_points = file.readlines()
    test_points = test_points[1:]
    x_Width = []
    y_Heigth = []
    
    for test_points_number in test_points:
        test_points_number = test_points_number[4:-2]
        print(test_points_number)
        
        x_Width.append(test_points_number.split(', ')[0])

        y_Heigth.append(test_points_number.split(', ')[1])

print(x_Width)
print(y_Heigth)

dist_list = []
for x1, y1 in zip (x_Width, y_Heigth):
    for Width, Height, Label in zip(PokemonWidth, PokemonHeigth, PokemonLabel):
        testW = 25
        testH = 32
        distance = math.sqrt(((testW - Width)**2) + (((testH - Height)**2)))
        dist_list.append([distance, Label])

print(dist_list)
dist_list.sort()



```
